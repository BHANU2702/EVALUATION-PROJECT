# Removing Outliers Using IQR Method

```python
# Calculating IQR

columns=['R','ERA','SHO','SV','E']
Q1=df[columns].quantile(0.25)
Q3=df[columns].quantile(0.75)
IQR=Q3 - Q1

# Determine outliers based on IQR
lower_value= Q1 - 1.5 * IQR
upper_value= Q3 + 1.5 * IQR

#Identify Outliers
outliers = df[((df[columns]<lower_value) | (df[columns]>upper_value)).any(axis=1)]

# Remove Outliers
new_df= df[~((df[columns]<lower_value) | (df[columns]>upper_value)).any(axis=1)]

# Prit cleaned new dataframe
print(" New Dataframe")
print(new_df)


```

```python
# Importing Libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

# Creating Train Test Split

```python
# Finding  the best random state
maxAccu = 0
maxRS = 0
for i in range(1,50):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=i)
    
    # Initiating and training the GradientBoostingRegressor
    GBR=GradientBoostingRegressor()
    GBR.fit(x_train,y_train)
    pred=GBR.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS= i
print("Maximum r2 score is ", maxAccu,"on Random_State",maxRS)            
```
